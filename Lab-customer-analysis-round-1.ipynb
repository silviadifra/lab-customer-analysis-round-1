```python
file3.rename(columns = {'state':'st'}, inplace = True)
print(file3.columns)
```

```python
file3 = file3[['customer', 'st', 'gender', 'education', 'customer lifetime value',
       'income', 'monthly premium auto', 'number of open complaints',
       'policy type', 'vehicle class', 'total claim amount']]
display(file3.head(3))
```

```python
#Concatenate the three dataframes
file4 = pd.concat([file1, file2, file3], axis=0) #discuss axis here: axis = 0 -> Merge vertically (by column)
print(file4.shape)
display(file4)
```

```python
#Which columns are numerical?Which columns are categorical??
file4.info()
#Income, monthly premium auto and total claim amount are numerical and all the rest are categorical
```

```python
#Perform the data cleaning operations mentioned so far in class
#Delete the column education and the number of open complaints from the dataframe.

file4 = file4.drop(['education','number of open complaints'], axis=1) # axis = 1 -> columns
file4.head()
```

```python
#Correct the values in the column customer lifetime value. 
#They are given as a percent, so multiply them by 100 and change dtype to numerical type.

def clean(x):
    if str(x)[-1]=='%':
        x = round(float(str(x).replace('%','')),4)/100
    else:
        x = round(float(x),2)
    return x
file4['customer lifetime value']=file4['customer lifetime value'].apply(clean)
display(file4.head())
file4.info()
```

```python
#Check for duplicate rows in the data and remove if any.
print(file4.shape)
file4 = file4.copy()
file4 = file4.drop_duplicates()
print(file4.shape)

```

```python
#Filter out the data for customers who have an income of 0 or less.
file5=file4[file4['income']<=0]
display(file5.head(10))
```
